```python
from collections import defaultdict
from dataclasses import dataclass
from io import StringIO
from pathlib import Path
import re

import numpy as np
import pandas as pd
import polars as pl
import polars.lazyframe as plf

DATA_DIR = Path('../data')
DATA_DIR.mkdir(exist_ok=True, parents=True)

RAW_DIR = DATA_DIR / 'raw'
RAW_DIR.mkdir(exist_ok=True)

PROCESSED_DIR = DATA_DIR / 'processed'
PROCESSED_DIR.mkdir(exist_ok=True)
```

```python
dataset_fps = list(RAW_DIR.glob('*.csv'))
print(f'Found {len(dataset_fps)} dataset files in {RAW_DIR}')

lazy_df = pl.concat([pl.scan_csv(fp) for fp in dataset_fps[:]])
lazy_df.count().collect(), lazy_df.head().collect()
```

```python
df = lazy_df.collect()
```

```python
merged_df = df.sort('clean_text').group_by('clean_text').agg(pl.col('len').sum().alias('len')).sort(['len', 'clean_text'], descending=[True, False])
merged_df
```

```python
merged_df.write_csv(DATA_DIR / 'processed' / 'wikipedia_russian_word_frequencies.csv', include_bom=True)
```
